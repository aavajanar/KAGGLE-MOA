### Train Features


#### Categorical features

```python
plt.style.use("seaborn")
sns.set_style('whitegrid')
fig = plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
sns.countplot(x='cp_type', data=train_features)
plt.title("cp_type on train set")

plt.subplot(1,3,2)
sns.countplot(x='cp_dose', data=train_features)
plt.title("cp_dose on train set")

plt.subplot(1,3,3)
sns.countplot(x='cp_time', data=train_features)
plt.title("cp_time on train set")
```

#### Numeric features

##### Gene data

```python
genes = [g for g in train_features.columns if g.startswith("g-")]
print(f"Number of g- features: {len(genes)}")

fig = plt.figure(figsize=(16,16))
choices = np.random.choice(len(genes), 16)
for i, col in enumerate(choices):
  gene_feature = "g-"+str(col)
  plt.subplot(4,4,i+1)
  sns.histplot(x=gene_feature, data=train_features)
```

##### Cell data

```python
cells = [c for c in train_features.columns if c.startswith("c-")]
print(f"Number of c- features: {len(cells)}")

fig = plt.figure(figsize=(16,16))
choices = np.random.choice(len(cells), 16)
for i, col in enumerate(choices):
  cell_feature = "c-"+str(col)
  plt.subplot(4,4,i+1)
  sns.histplot(x=cell_feature, data=train_features)
```

#### Correlations

##### Cell correlations


```python
# only looking at treated samples
cellsdf = train_features[train_features['cp_type'] == 'trt_cp'][cells]

# look at the highest correlation values
cells_corr = cellsdf.corr()
np.fill_diagonal(cells_corr.values, np.nan)
cells_corr = cells_corr.abs().unstack().sort_values(kind="quicksort", ascending=False)
print(cells_corr.head())

# visualize the correlations
fig = plt.figure(figsize=(25,10))
sns.heatmap(cellsdf.corr())
```

##### Gene correlations

```python
# only looking at treated samples
genesdf = train_features[train_features['cp_type'] == 'trt_cp'][genes]

# look at the highest correlation values
genes_corr = genesdf.corr()
np.fill_diagonal(genes_corr.values, np.nan)
genes_corr = genes_corr.abs().unstack().sort_values(kind="quicksort", ascending=False)
print("Highest values:")
print(genes_corr.head())

# visualize the correlations
fig = plt.figure(figsize=(25,10))
sns.heatmap(genesdf.corr())
```
